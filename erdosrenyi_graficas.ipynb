# Erdös-Renyi: Gráficas aleatorias 

```python
import statistics
import random
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import collections
```

Definimos la clase Vertice a partir de la cual generaremos la gráfica de Erdos-Renyi

```python
class Vertice:
   def __init__(self, index):
      self.index = index
      self.neighbors = []
 
   def __repr__(self):
      return repr(self.index)
```

Esta función genera la gráfica de ER. Recibe como entrada $n$ y $p$. Regresa como salida una lista de vértices y una lista de aristas

```python
def erdosRenyi(n,p):
   vertices = [Vertice(i) for i in range(n)]
   aristas = [(i,j) for i in range(n) for j in range(i) if random.random() < p]
 
   for (i,j) in aristas:
      vertices[i].neighbors.append(vertices[j])
      vertices[j].neighbors.append(vertices[i])
 
   return vertices,aristas
```

Generamos una primera gráfica de ER.

```python
grafica1=erdosRenyi(50,0.25)
```

```python
grafica1[1]
```

Convertimos la lista de aristas en una gráfica de networkx.

```python
G = nx.Graph()  
G.add_edges_from(grafica1[1])
```
