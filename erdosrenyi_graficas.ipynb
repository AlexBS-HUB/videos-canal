# Erdös-Renyi: Gráficas aleatorias 

```python
import statistics
import random
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python
class Node:
   def __init__(self, index):
      self.index = index
      self.neighbors = []
 
   def __repr__(self):
      return repr(self.index)
```

Esta función genera la gráfica de ER. Recibe como entrada $n$ y $p$. Regresa como salida una lista de vértices y una lista de aristas

```python
def erdosRenyi(n,p):
   nodes = [Node(i) for i in range(n)]
   edges = [(i,j) for i in range(n) for j in range(i) if random.random() < p]
 
   for (i,j) in edges:
      nodes[i].neighbors.append(nodes[j])
      nodes[j].neighbors.append(nodes[i])
 
   return nodes,edges
```

Generamos una primera gráfica de ER.

```python
grafica1=erdosRenyi(50,0.25)
```

```python
aristas = grafica1[1]
```

Convertimos la lista de aristas en una gráfica de networkx.

```python
G = nx.Graph()  
G.add_edges_from(aristas)
```

```python
nx.info(G)
```

```python
plt.figure(figsize=(18,12))
nx.draw_circular(G)
```
