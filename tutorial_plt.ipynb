# Tutorial Matplotlib (Pokemon Dataset)

Importamos Matplotlib y Pandas.

%matplotlib notebook -> plots interactivos 

%matplotlib inline -> plots estáticos

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

Ajustamos el formato

```python
plt.style.use('ggplot')
```

Convertimos el archivo .csv en un DataFrame

```python
poke_lista = pd.read_csv('pokemon.csv', index_col=None)
poke_lista
```

```python
poke_lista.iloc[175]
```

Reemplazamos las celdas en blanco, NaN, por ceros

```python
poke_lista = poke_lista.fillna(0)
```

```python
poke_lista.iloc[100]
```

Cada renglón de este DataFrame corresponde con un pokemon

Nos concentraremos en la primera generación de Pokemon.

```python
gen1 = poke_lista.loc[0:149]
gen1
```

## Grafiquemos ahora con Matplotlib a partir de un DataFrame

## Gráfica de Barras

```python
gen1.plot(x='Name', y='HP', kind='bar',figsize=(20, 6)) ### Similar a Matlab
```

Como son muchos, nos enfocaremos en los que son tipo insecto.

```python
insecto = gen1.loc[gen1['Type 1'] == 'Bug']
insecto
```

```python
insecto.plot(x='Name', y='HP', kind='bar')
```

Etiquetamos los ejes y modificamos el tamaño

```python
insecto.plot(x='Name', y='HP', kind='bar',figsize=(12, 6))

plt.title("Pokemon tipo Bicho generación 1")
plt.xlabel("Pokemon")
plt.ylabel("HP")
```

Rotamos las etiquetas del eje x

```python
insecto.plot(x='Name', y='HP', kind='bar',figsize=(12, 6))

plt.xticks(rotation=45, horizontalalignment="center")
plt.title("Pokemon tipo Bicho generación 1")
plt.xlabel("Pokemon")
plt.ylabel("HP")
```

Podemos modificar el color

```python
insecto.plot(x='Name', y='HP', kind='bar',color='c',figsize=(12, 6))

plt.xticks(rotation=60, horizontalalignment="center")
plt.title("Pokemon tipo Bicho generación 1")
plt.xlabel("Pokemon")
plt.ylabel("HP")
```

Gráfico de barras horizontal

```python
insect_ord = insecto.sort_values(by=['HP']) ### Ordenamos de manera descendente
```

```python
insect_ord.plot(x='Name', y='HP', kind='barh',figsize=(6, 8),color='m')

plt.xticks(rotation=60, horizontalalignment="center")
plt.title("Gráfica de Barras en Matplotlib")
plt.xlabel("Hit Points")
plt.ylabel("Pokemon")
```

Hagamos ahora un comparativo de Ataque y Defensa de cada Pokemon

```python
### Podemos manipular la anchura de las barras
insecto.plot(x='Name', y=['Attack','Defense'], kind='bar',figsize=(12,6),color=['orange','lightblue'],width=0.8)
```

Agregamos los HP

```python
### Podemos manipular la anchura de las barras
insecto.plot(x='Name', y=['HP','Attack','Defense'], kind='bar',figsize=(12,6),color=['m','orange','lightblue'],width=0.8)
```

Revisemos ahora cuales son los 10 con mayor ataque

```python
first_attack = gen1.nlargest(10, 'Attack')
first_attack
```

```python
first_attack.plot(x='Name', y=['HP','Attack','Defense'], kind='bar',figsize=(12,6),color=['m','orange','lightblue'],width=0.8)
```

## Tipos de Pokemon y gráfico de Pastel

Contamos cuantos pokemon hay de cada tipo:

```python
from collections import Counter

c = Counter(list(zip(gen1['Type 1'])))
type(c)
```

```python
traduccion = ['Planta','Fuego','Agua','Insecto','Normal','Venenoso','Eléctrico','Tierra','Hada','Lucha','Psíquico',
              'Roca','Fantasma','Fuego','Dragón']
```

Creamos un DataFrame con esta información

```python
lista = [list(i) for i in c.items()]
tipos_pokemon = pd.DataFrame(lista,index= traduccion, columns =['Tipo','Cantidad'])
```

Ordenamos el DataFrame de mayor a menor

```python
tipos_pokemon = tipos_pokemon.sort_values(by=['Cantidad'],ascending=False)
tipos_pokemon
```

Graficamos el anterior DataFrame

```python
grafica = tipos_pokemon.plot.pie(y='Cantidad', figsize=(16, 10),autopct='%1.2f%%')
grafica.get_legend().remove()
```

Quitamos las leyendas sobre las partes del pastel

```python
grafica = tipos_pokemon.plot.pie(y='Cantidad', figsize=(16, 10),autopct='%1.2f%%',labels=['','','','','','','','','','','','','','',''])
grafica.legend(traduccion)
```

Podemos cambiar el lugar de la cajita

```python
grafica = tipos_pokemon.plot.pie(y='Cantidad', title= 'Tipos de Pokemon',figsize=(16, 10),
           autopct='%1.2f%%',labels=['','','','','','','','','','','','','','',''])
grafica.legend(traduccion,loc='upper left',ncol=2)
```

Ajustamos el tamaño de la cajita para que no tape nuestra gráfica

```python
from matplotlib.font_manager import FontProperties

fontP = FontProperties()
fontP.set_size('x-small')
```

Además guardamos nuestra imagen en formato jpg

```python
grafica = tipos_pokemon.plot.pie(y='Cantidad', title= 'Tipos de Pokemon', 
          figsize=(10, 10),autopct='%1.2f%%',labels=['','','','','','','','','','','','','','',''])
grafica.legend(traduccion,loc='upper left',ncol=2,prop=fontP)

### Guardaremos esta gráfica
plt.savefig('mi_grafica.jpg')
```
